```python
from jericho import *
import glob
import spacy
import json
import nltk
from nltk.corpus import verbnet
from nltk.corpus import wordnet as wn
nltk.download('verbnet')
nltk.download('wordnet');
```

```python
walkthroughs = []
games = glob.glob("z-machine-games-master/jericho-game-suite/*.*")
for game in games:
  print(game)
  env = FrotzEnv(game)
  walkthrough = env.get_walkthrough()
  if walkthrough is not None and walkthrough != []:
    print(walkthrough)
    walkthrough = [action.lstrip("[").rstrip("]").strip().lower() for action in walkthrough if action != ""]
    walkthrough = sorted(set(walkthrough))
    walkthroughs.append((game, walkthrough))
```

```python
print(walkthroughs)
```

```python
l = sorted(set(["upload image", "download image", "upload picture", "upload video", "download video", "save image", "save mp4", "surf the net", "launch command", "hack terminal", "record scene", "take a picture", "take selfie", "shoot video", "call Maxx", "open app", "open terminal", "search picture of cat", "google it"]))
walkthroughs.append(("", l))
print(walkthroughs[-1])
```

```python
hyper = lambda s: s.hypernyms()
abstr = wn.synset('abstraction.n.06')
nlp = spacy.load("en_core_web_sm")
with open("explosion.txt", "w") as e:
  for game, actions in walkthroughs:
    for action in actions:
      doc = nlp(action)
      dic = {}
      dic["action"] = doc.text
      dic["game"] = game
      for token in doc:
        i = token.i
        dic[i] = {}
        dic[i]["len"] = len(token)
        dic[i]["text"] = token.lower_
        dic[i]["norm"] = token.norm_
        dic[i]["pos"] = token.pos_
        dic[i]["tag"] = token.tag_
        dic[i]["dependency"] = token.dep_
        dic[i]["dep_parent"] = token.head.i
        dic[i]["digit"] = token.is_digit
        dic[i]["shape"] = token.shape_
        dic[i]["wordnet"] = [(str(synset), synset.offset(), [l.key() for l in synset.lemmas()], synset.definition(), synset.examples()[0] if len(synset.examples()) > 0 else "", abstr in list(synset.closure(hyper))) for synset in wn.synsets(token.norm_)]
        dic[i]["wn_offset"] = ""
        dic[i]["wn_isabs"] = ""
        dic[i]["verbnet"] = [(verb, verbnet.wordnetids(verb)) for verb in verbnet.classids(token.norm_)]
      e.write(json.dumps(dic, indent=4))
      e.write("\n")
```
