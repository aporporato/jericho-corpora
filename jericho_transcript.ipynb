```python
walkthroughs = []
games = glob.glob("z-machine-games-master/jericho-game-suite/*.*")
for i, game in enumerate(games):
  print(i, game)
  env = FrotzEnv(game)
  walkthrough = env.get_walkthrough()
  if walkthrough is not None and walkthrough != []:
    env.step('script')
    for act in walkthrough:
      env.step(act)
    env.step('unscript')
```
