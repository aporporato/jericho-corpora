```python
actions = []
```

```python
from jericho import *
import glob
games = glob.glob("z-machine-games-master/jericho-game-suite/*.*")
for game in games:
  print(game)
  env = FrotzEnv(game)
  walkthrough = env.get_walkthrough()
  if walkthrough is not None and walkthrough != []:
    print(walkthrough)
    actions.extend([a.lstrip("[").rstrip("]").strip().lower() for a in walkthrough])
```

```python
print(actions)
```

```python
actions = set(actions).union({"upload image", "download image", "upload picture", "upload video", "download video", "save image", "save mp4", "surf the net", "launch command", "hack terminal", "record scene", "take a picture", "take selfie", "shoot video", "call Maxx", "open app", "open terminal", "search picture of cat", "google it", "watch video"})
actions = [action.strip() for action in actions if action != ""]
actions = sorted(actions)
```

```python
with open("actions_stripped_sed_sorted_aug.txt", "w") as f:
  f.writelines([action.strip() + "\n" for action in actions])
```
